# Environment Check

Dieses Notebook prueft die Ausfuehrungsumgebung: gemountete Projektordner, Kafka-Erreichbarkeit und SparkSession. Es gehoert nicht zur nummerierten Analyse-Pipeline aus BD-03.

```python
from project_paths import project_root as get_project_root

base_path = get_project_root()
expected_dirs = ['notebooks', 'data', 'outputs']

for directory in expected_dirs:
    path = base_path / directory
    print(f'{path}: exists={path.exists()}')
    assert path.exists(), f'Missing mounted directory: {path}'
```

```python
import os
import socket

bootstrap_servers = os.environ.get('KAFKA_BOOTSTRAP_SERVERS', 'kafka:29092')
host, port = bootstrap_servers.split(':')

with socket.create_connection((host, int(port)), timeout=10):
    print(f'Kafka reachable at {bootstrap_servers}')
```

```python
import os

from project_paths import ensure_pyspark_path

ensure_pyspark_path()

from pyspark.sql import SparkSession

spark_master = os.environ.get('SPARK_MASTER', 'local[*]')

spark = (
    SparkSession.builder
    .appName('football-weather-environment-check')
    .master(spark_master)
    .getOrCreate()
)

print(f'Spark master: {spark_master}')
print(f'Spark version: {spark.version}')
spark.range(5).show()
spark.stop()
```
